```python
import numpy as np
from scipy.stats import beta

import matplotlib.pyplot as plt
```

```python
# choose the prior of A

a = 1
b = 1

# choose the prior of B

c = 100
d = 100

# plot the priors

x = np.linspace(0,1,100)
y = beta.pdf(x,a,b)

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("Prob")
plt.title("Prior distribution of A")
plt.show()

x = np.linspace(0,1,100)
y = beta.pdf(x,c,d)

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("Prob")
plt.title("Prior distribution of B")
plt.show()
```

```python
# coin flip with unknown bias
bias = np.random.beta(a=1,b=1)
coin = np.random.binomial(n=1, p=bias, size=100)

print(bias)
print(coin)
```

```python
# two bayesian learners A and B with beta priors observe the coin and update by conditionalization

for x in coin:
    # print (x)
    if x == 1:
      a = a + 1
    if x == 0:
      b = b + 1

for x in coin:
    # print (x)
    if x == 1:
      c = c + 1
    if x == 0:
      d = d + 1

# final predictive probabilities

print(a/(a+b))
print(c/(c+d))

# plot the posteriors of A and B

x = np.linspace(0,1,100)
y = beta.pdf(x,a,b)

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("Prob")
plt.title("Posterior distribution of A")
plt.show()

x = np.linspace(0,1,100)
y = beta.pdf(x,c,d)

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("Prob")
plt.title("Posterior distribution of B")
plt.show()
```
